```python
import time


start = time.time()


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def catalogo_filmes(df_title_basics, df_title_ratings):
    if 'tconst' not in df_title_basics.columns or 'tconst' not in df_title_ratings.columns:
        print("Erro: O Código do filme não foi encontrado nos índices.")
        return None

    df_filmes = pd.merge(df_title_basics, df_title_ratings, on='tconst', how='inner')
    filmes_recentes = df_filmes[df_filmes['startYear'].fillna(0).astype(int) > 2017]
    
    if len(filmes_recentes) < 20:
        print("Ajuste a pesquisa. Por favor, a quantidade de filmes desejada.")
        return None

    filmes_aleatorios = filmes_recentes.sample(20, replace=False)

    catalogo = {}
    for idx, filme in filmes_aleatorios.iterrows():
        catalogo[idx] = {
            "titulo": filme['primaryTitle'],
            "ano": str(int(filme['startYear'])),
            "genero": filme['genres'] if pd.notna(filme['genres']) else '',
            "preco": 5.00,
            "avaliacao": filme['averageRating'],
            "diaria": 5.00
        }

    return catalogo

def locacao_filmes(catalogo):
    if catalogo is None:
        return 0, {}, []

    total_locacao = 0.0
    filmes_locados = {}
    avaliacoes = {}
    generos = []

    print("\n----------------------------------------------------------------------------------------------------------------------------------")
    print("|                                    Catálogo de Filmes Atualizado                                                                 |")
    print("----------------------------------------------------------------------------------------------------------------------------------")

    for codigo, filme in catalogo.items():
        titulo = filme['titulo'][:20] + '...' if len(filme['titulo']) > 20 else filme['titulo']
        genero = str(filme['genero'])[:20] + '...' if len(str(filme['genero'])) > 20 else str(filme['genero'])
        print(f"| {codigo:<7} | {titulo:<25} | Ano: {filme['ano']:<10} | Gênero: {genero:<24} | Avaliação: {filme['avaliacao']:<6} | Diária: R$ {filme['diaria']:<6} |")
        if filme['genero']:
            generos.extend(filme['genero'].split(','))

    print("------------------------------------------------------------------------------------------------------------------------------------")

    while True:
        codigo_filme = input("Selecione o código do filme para locação (ou 0 para encerrar): ")
        if codigo_filme == "0":
            break
        elif codigo_filme.isdigit() and int(codigo_filme) in catalogo:
            codigo_filme = int(codigo_filme)
            print(f"Valor da diária por filme: R$ {catalogo[codigo_filme]['diaria']:.2f}")
            quantidade_dias = int(input("Digite a quantidade de dias para locação: "))
            filme = catalogo[codigo_filme]
            preco_locacao = filme["preco"] * quantidade_dias
            total_locacao += preco_locacao
            filmes_locados[codigo_filme] = filme
        else:
            print("Opção inválida. Tente novamente.")
            continue
            
    print('-' * 70)
    print("\nTotal a pagar: R$ {:.2f}".format(total_locacao))
    print("Locação encerrada. Obrigado por usar nosso serviço!")
    print("Por favor, retorne para avaliar os filmes locados. Sua opinião é importante para nós!")
    print('-' * 70)

    for codigo, filme in filmes_locados.items():
        avaliacao = input(f"De 0 a 10, qual a sua avaliação para '{filme['titulo']}'? ")
        while not avaliacao.isdigit() or int(avaliacao) < 0 or int(avaliacao) > 10:
            print("Avaliação inválida. A avaliação deve ser um número de 0 a 10.")
            avaliacao = input(f"De 0 a 10, qual a sua avaliação para '{filme['titulo']}'? ")
        avaliacoes[codigo] = int(avaliacao)

    print("\nAvaliações dos Filmes:")
    for codigo, nota in avaliacoes.items():
        filme = catalogo[codigo]
        print(f"{filme['titulo']} - Avaliação: {nota}")
        print('-' * 70)

    return total_locacao, avaliacoes, generos



if not df_title_basics.empty and not df_title_ratings.empty:
    catalogo = catalogo_filmes(df_title_basics, df_title_ratings)

if catalogo is not None:
    total_locacao, avaliacoes, generos = locacao_filmes(catalogo)
else:
    total_locacao = 0
    avaliacoes = {}
    generos = []


if generos:
    generos_counts = pd.Series(generos).value_counts()
    plt.figure(figsize=(14, 8))
    plt.pie(generos_counts, labels=generos_counts.index, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 12})
    plt.title('Distribuição dos Gêneros dos Filmes no Catálogo', fontsize=18)
    plt.axis('equal')
    plt.show()


if avaliacoes:
    avaliacoes_df = pd.DataFrame({
        'Código do Filme': list(avaliacoes.keys()),
        'Avaliação': list(avaliacoes.values())
    })
    avaliacoes_df['Título'] = avaliacoes_df['Código do Filme'].apply(lambda x: catalogo[x]['titulo'])
    avaliacoes_df['Gênero'] = avaliacoes_df['Código do Filme'].apply(lambda x: catalogo[x]['genero'])

    plt.figure(figsize=(14, 8))
    sns.barplot(x='Título', y='Avaliação', hue='Gênero', data=avaliacoes_df, palette='viridis', dodge=False)
    plt.title('Avaliações dos Filmes Locados', fontsize=18)
    plt.xlabel('Título do Filme', fontsize=14)
    plt.ylabel('Avaliação', fontsize=14)
    plt.xticks(rotation=45, fontsize=12, ha='right')
    plt.yticks(fontsize=12)
    plt.legend(title='Gênero', fontsize=12)
    for index, row in avaliacoes_df.iterrows():
        plt.text(index, row['Avaliação'] + 0.1, f"{row['Avaliação']}", ha='center', fontsize=12)
    plt.tight_layout()
    plt.show()

end = time.time()
tempo_execucao = end - start
print(f"O tempo de execução do código foi de {tempo_execucao} segundos.")
```

### 12. Faça uma análise dos filmes e mostre para o investidor uma análise gerencial dos filmes no dataset (análise livre para mostrar sua criatividade, desenvolvimento e habilidades do Python).


- ### Criando Sessão Spark, importações sem otimizar com parametros, tempo execução 42.122702836990356 segundos.

- Arquivos .tsv - código alterado para funções PySpark

- Utilizando PySpark sem tratamento nos arquivos e executando na mesma célula não foi performático

<br>

```python
import time


start = time.time()


from pyspark.sql import SparkSession
from pyspark.sql import functions as F


spark = SparkSession.builder \
    .appName('Catalogo de Filmes') \
    .getOrCreate()


path_title_basics = r"C:\Users\Thiago10\OneDrive\Área de Trabalho\ARQ_JUMP_START\MODULOPYTHONJUMP\CASEPYTHON\title.basics.tsv\title.basics.tsv"
path_title_ratings = r"C:\Users\Thiago10\OneDrive\Área de Trabalho\ARQ_JUMP_START\MODULOPYTHONJUMP\CASEPYTHON\title.ratings.tsv\title.ratings.tsv"

try:
   
    df_title_basics = spark.read.csv(path_title_basics, header=True, sep='\t', inferSchema=True)
    df_title_ratings = spark.read.csv(path_title_ratings, header=True, sep='\t', inferSchema=True)

    
    df_title_basics = df_title_basics.na.replace('\\N', None)
    df_title_ratings = df_title_ratings.na.replace('\\N', None)

    
    print("Primeiras 3 linhas do DataFrame Title Basics:")
    df_title_basics.show(3)
    
    print("Primeiras 3 linhas do DataFrame Title Ratings:")
    df_title_ratings.show(3)

except Exception as e:
    print(f"Erro na leitura dos arquivos: {e}")
    
# Função para catalogar filmes
def catalogo_filmes(df_title_basics, df_title_ratings):
    
    # Valores de configuração
    LIMIT_FILMES = 20
    DIARIA = 5.00

    # Checando chaves
    if 'tconst' not in df_title_basics.columns or 'tconst' not in df_title_ratings.columns:
        print("Erro: O código do filme não foi encontrado nos índices.")
        return None

    # Mesclando os DataFrames e filtrando filmes lançados após 2017
    df_filmes = df_title_basics.join(df_title_ratings, on='tconst', how='inner') \
        .filter((F.col('startYear').isNotNull()) & (F.col('startYear').cast('int') > 2017))

    # Verifica se existem filmes suficientes
    if df_filmes.count() < LIMIT_FILMES:
        print("Ajuste a pesquisa. Por favor, a quantidade de filmes desejada.")
        return None

    # Seleciona aleatoriamente 20 filmes
    filmes_aleatorios = df_filmes.orderBy(F.rand()).limit(LIMIT_FILMES)

    # Coleta as informações em um formato de dicionário
    catalogo = filmes_aleatorios.select(
        F.col('primaryTitle').alias('titulo'), 
        F.when(F.col('startYear').isNotNull(), F.col('startYear').cast('int')).cast('string').alias('ano'),
        F.col('genres').alias('genero'),
        F.lit(DIARIA).alias('diaria'),
        F.col('averageRating').alias('avaliacao')
    ).collect()

    # Convertendo o resultado em dicionário
    catalogo_dict = {idx + 1: {
        "titulo": filme['titulo'],
        "ano": filme['ano'] if filme['ano'] is not None else '',
        "genero": filme['genero'] if filme['genero'] is not None else '',
        "preco": DIARIA,
        "avaliacao": filme['avaliacao'],
        "diaria": DIARIA
    } for idx, filme in enumerate(catalogo)}

    return catalogo_dict


def locacao_filmes(catalogo):
    if catalogo is None:
        return 0, {}, []

    total_locacao = 0.0
    filmes_locados = {}
    avaliacoes = {}
    generos = []

    print("\n----------------------------------------------------------------------------------------------------------------------------------")
    print("|                                    Catálogo de Filmes Atualizado                                                                 |")
    print("----------------------------------------------------------------------------------------------------------------------------------")

    
    for codigo, filme in catalogo.items():
        titulo = filme['titulo'][:20] + '...' if len(filme['titulo']) > 20 else filme['titulo']
        genero = str(filme['genero'])[:20] + '...' if len(str(filme['genero'])) > 20 else str(filme['genero'])
        print(f"| {codigo:<7} | {titulo:<25} | Ano: {filme['ano']:<10} | Gênero: {genero:<24} | Avaliação: {filme['avaliacao']:<6} | Diária: R$ {filme['diaria']:<6} |")
        if filme['genero']:
            generos.extend(filme['genero'].split(','))

    print("------------------------------------------------------------------------------------------------------------------------------------")

    while True:
        codigo_filme = input("Selecione o código do filme para locação (ou 0 para encerrar): ")
        if codigo_filme == "0":
            break
        elif codigo_filme.isdigit() and int(codigo_filme) in catalogo:
            codigo_filme = int(codigo_filme)
            print(f"Valor da diária por filme: R$ {catalogo[codigo_filme]['diaria']:.2f}")
            quantidade_dias = int(input("Digite a quantidade de dias para locação: "))
            filme = catalogo[codigo_filme]
            preco_locacao = filme["preco"] * quantidade_dias
            total_locacao += preco_locacao
            filmes_locados[codigo_filme] = filme
        else:
            print("Opção inválida. Tente novamente.")
            continue
            
    print('-' * 70)
    print("\nTotal a pagar: R$ {:.2f}".format(total_locacao))
    print("Locação encerrada. Obrigado por usar nosso serviço!")
    print("Por favor, retorne para avaliar os filmes locados. Sua opinião é importante para nós!")
    print('-' * 70)

    for codigo, filme in filmes_locados.items():
        avaliacao = input(f"De 0 a 10, qual a sua avaliação para '{filme['titulo']}'? ")
        while not avaliacao.isdigit() or int(avaliacao) < 0 or int(avaliacao) > 10:
            print("Avaliação inválida. A avaliação deve ser um número de 0 a 10.")
            avaliacao = input(f"De 0 a 10, qual a sua avaliação para '{filme['titulo']}'? ")
        avaliacoes[codigo] = int(avaliacao)

    print("\nAvaliações dos Filmes:")
    for codigo, nota in avaliacoes.items():
        filme = catalogo[codigo]
        print(f"{filme['titulo']} - Avaliação: {nota}")
        print('-' * 70)

    return total_locacao, avaliacoes, generos

# Executando a função principal
if df_title_basics.count() > 0 and df_title_ratings.count() > 0:
    catalogo = catalogo_filmes(df_title_basics, df_title_ratings)

if "catalogo" in locals() and catalogo is not None:
    total_locacao, avaliacoes, generos = locacao_filmes(catalogo)
else:
    total_locacao = 0
    avaliacoes = {}
    generos = []

# Exibir resultados finais com formatação
print(f'Total de locação: R$ {total_locacao:.2f}')
print("\nAvaliações:")
if avaliacoes:
    for codigo, nota in avaliacoes.items():
        titulo = catalogo[codigo]['titulo']
        print(f"{titulo} - Avaliação: {nota}")
else:
    print("Nenhuma avaliação foi registrada.")

print("\nGêneros:")
if generos:
    generos_set = set(generos)  # Usar um conjunto para evitar duplicatas
    for genero in generos_set:
        print(genero)
else:
    print("Nenhum gênero foi registrado.")

end = time.time()
tempo_execucao = end - start
print(f"O tempo de execução do código foi de {tempo_execucao} segundos.")
```

- ### Com funções spark.sql com arquivos importados .parquet melhorou o tempo em relação Pyspark puro 6.664385795593262 segundos.

```python
import time

# Medindo o tempo de execução do código
start = time.time()

from pyspark.sql import SparkSession
from pyspark.sql import functions as F

def catalogo_filmes(df_title_basics, df_title_ratings):

    # Registrando os DataFrames como tabelas temporárias
    df_title_basics.createOrReplaceTempView("title_basics")
    df_title_ratings.createOrReplaceTempView("title_ratings")
    
    # Consulta SQL para mesclar e filtrar os filmes lançados após 2017
    df_filmes = spark.sql(f"""
               SELECT 
                      tb.primaryTitle as titulo, 
                      CAST(tb.startYear AS STRING) as ano, 
                      tb.genres as genero, 
                      5.00 as diaria, 
                      tr.averageRating as avaliacao
               FROM title_basics tb
               JOIN title_ratings tr ON tb.tconst = tr.tconst
               WHERE tb.startYear IS NOT NULL AND CAST(tb.startYear AS INT) > 2017
               ORDER BY rand()
               LIMIT 20
    """)

    LIMIT_FILMES = 20

    filmes_aleatorios = df_filmes.orderBy(F.rand()).limit(LIMIT_FILMES)

    DIARIA = 5.00

    catalogo = filmes_aleatorios.select(
        F.col('titulo'), 
        F.when(F.col('ano').isNotNull(), F.col('ano').cast('int')).cast('string').alias('ano'),
        F.col('genero').alias('genero'),
        F.lit(DIARIA).alias('diaria'),
        F.col('avaliacao').alias('avaliacao')
    ).collect()

    catalogo_dict = {}
    
    for idx, filme in enumerate(catalogo):
        catalogo_dict[idx + 1] = {
            "titulo": filme['titulo'],
            "ano": filme['ano'] if filme['ano'] is not None else '',
            "genero": filme['genero'] if filme['genero'] is not None else '',
            "preco": DIARIA,
            "avaliacao": filme['avaliacao'],
            "diaria": DIARIA
        }

  
    return catalogo_dict

def locacao_filmes(catalogo):
    if catalogo is None:
        return 0, {}, []

    total_locacao = 0.0
    filmes_locados = {}
    avaliacoes = {}
    generos = []

    print("\n----------------------------------------------------------------------------------------------------------------------------------")
    print("|                                    Catálogo de Filmes Atualizado                                                                 |")
    print("----------------------------------------------------------------------------------------------------------------------------------")

    for codigo, filme in catalogo.items():
        titulo = filme['titulo'][:20] + '...' if len(filme['titulo']) > 20 else filme['titulo']
        genero = str(filme['genero'])[:20] + '...' if len(str(filme['genero'])) > 20 else str(filme['genero'])
        print(f"| {codigo:<7} | {titulo:<25} | Ano: {filme['ano']:<10} | Gênero: {genero:<24} | Avaliação: {filme['avaliacao']:<6} | Diária: R$ {filme['diaria']:<6} |")

        if filme['genero']:
            generos.extend(filme['genero'].split(','))

    print("------------------------------------------------------------------------------------------------------------------------------------")

    while True:
        codigo_filme = input("Selecione o código do filme para locação (ou 0 para encerrar): ")
        if codigo_filme == "0":
            break
        elif codigo_filme.isdigit() and int(codigo_filme) in catalogo:
            codigo_filme = int(codigo_filme)
            print(f"Valor da diária por filme: R$ {catalogo[codigo_filme]['diaria']:.2f}")
            quantidade_dias = int(input("Digite a quantidade de dias para locação: "))
            filme = catalogo[codigo_filme]
            preco_locacao = filme["preco"] * quantidade_dias
            total_locacao += preco_locacao
            filmes_locados[codigo_filme] = filme
        else:
            print("Opção inválida. Tente novamente.")
            continue
            
    print('-' * 70)
    print("\nTotal a pagar: R$ {:.2f}".format(total_locacao))
    print("Locação encerrada. Obrigado por usar nosso serviço!")
    print("Por favor, retorne para avaliar os filmes locados. Sua opinião é importante para nós!")
    print('-' * 70)

    for codigo, filme in filmes_locados.items():
        avaliacao = input(f"De 0 a 10, qual a sua avaliação para '{filme['titulo']}'? ")
        while not avaliacao.isdigit() or int(avaliacao) < 0 or int(avaliacao) > 10:
            print("Avaliação inválida. A avaliação deve ser um número de 0 a 10.")
            avaliacao = input(f"De 0 a 10, qual a sua avaliação para '{filme['titulo']}'? ")
        avaliacoes[codigo] = int(avaliacao)

    print("\nAvaliações dos Filmes:")
    for codigo, nota in avaliacoes.items():
        filme = catalogo[codigo]
        print(f"{filme['titulo']} - Avaliação: {nota}")
        print('-' * 70)

    return total_locacao, avaliacoes, generos

# Executando a função principal
catalogo = catalogo_filmes(df_title_basics, df_title_ratings)

if catalogo is not None:
    total_locacao, avaliacoes, generos = locacao_filmes(catalogo)
else:
    total_locacao = 0
    avaliacoes = {}
    generos = []


print(f'Total de locação: R$ {total_locacao:.2f}')
print("\nAvaliações:")
if avaliacoes:
    for codigo, nota in avaliacoes.items():
        titulo = catalogo[codigo]['titulo']
        print(f"{titulo} - Avaliação: {nota}")
else:
    print("Nenhuma avaliação foi registrada.")

print("\nGêneros:")
if generos:
    generos_set = set(generos)  # Usar um conjunto para evitar duplicatas
    for genero in generos_set:
        print(genero)
else:
    print("Nenhum gênero foi registrado.")
    
# Medindo o tempo de execução do código
end = time.time()
tempo_execucao = end - start
print(f"O tempo de execução do código foi de {tempo_execucao} segundos.")
```

```python
import time
start = time.time()

from pyspark.sql import SparkSession
from pyspark.sql import functions as F

def filmes_por_pais(spark, df_title_basics, df_title_akas):
    df_mesclar = df_title_basics.join(df_title_akas, df_title_basics['tconst'] == df_title_akas['titleId'])\
                                .filter((df_title_basics['titleType'] == 'movie') | (df_title_basics['titleType'] == 'short'))\
                                .groupBy('region')\
                                .agg(F.count('tconst').alias('Número de Filmes'))\
                                .select('region', 'Número de Filmes')\
                                .orderBy(F.desc('Número de Filmes'))
    
    return df_mesclar

filmes_pais = filmes_por_pais(spark, df_title_basics, df_title_akas)


filmes_pais.show(10)


end = time.time()
tempo_execucao = end - start
print(f"O tempo de execução do código foi de {tempo_execucao} segundos.")
```

```python
import time

start = time.time()

from pyspark.sql import SparkSession
from pyspark.sql import functions as F


def filmes_por_pais(spark, df_title_basics, df_title_akas):
    
    df_title_basics.createOrReplaceTempView("title_basics")
    df_title_akas.createOrReplaceTempView("title_akas")

    
    df_mesclar = spark.sql(f"""
        SELECT 
            COALESCE(ta.region, 'Desconhecido') AS region, 
            COUNT(tb.tconst) AS num_filmes
        FROM title_basics tb
        JOIN title_akas ta ON tb.tconst = ta.titleId
        WHERE tb.titleType IN ('movie', 'short')
        GROUP BY ta.region
        ORDER BY num_filmes DESC
    """)

    return df_mesclar



filmes_pais = filmes_por_pais(spark, df_title_basics, df_title_akas)


filmes_pais.show(10)


end = time.time()
tempo_execucao = end - start
print(f"O tempo de execução do código foi de {tempo_execucao} segundos.")
```
